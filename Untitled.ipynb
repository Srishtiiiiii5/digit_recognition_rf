```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
```

```python
dig= load_digits()
```

```python
dir(dig)
```

```python
plt.gray()
for i in range(4):
    plt.matshow(dig.images[i])
```

```python
df= pd.DataFrame(dig.data)
```

```python
df.head()
```

```python
dig.target
```

```python
df['target']= dig.target
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier  #ensemble used when we are using multiple algos
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis='columns'), dig.target, test_size=0.2)
```

```python
len(X_train)
```

```python
model= RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted= model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_predicted)
```

```python
cm
```
